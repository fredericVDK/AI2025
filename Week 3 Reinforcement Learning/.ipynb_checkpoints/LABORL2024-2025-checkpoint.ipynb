### 3. Implementeer een Q-learning agent in Python (4 ptn)

Schrijf nu ook een klasse __Agent__. In deze klasse voorzie je volgende velden en methoden
:
- __een discount factor__  $\gamma$ om je reward te berekenen. Een pad waarbij de oplossing sneller bereikt wordt is immers te prefereren. Initialiseer deze op een zelfgekozen waarde.

- velden om een __average\_reward__ bij te houden en een __episode\_reward__. Initialiseer beide op 0.

- een veld om het __aantal steps__ bij te houden in de huidige episode.

- een methode __select\_action__ die gegeven de huidige toestand van de omgeving een nieuwe actie selecteert en deze ook teruggeeft. Als je de agent initieel wil testen kan je hem een random actie laten kiezen. Wanneer je de agent wil laten leren kan je de $\epsilon$-greedy actie selectie strategie of de soft-max actie selectie strategie implementeren.

- een methode __step__ die gegeven een omgeving en een actie, deze actie uitvoert in de omgeving. De verkregen immediate reward verwerk je in de __average\_reward__. 

- een methode __clear\_episode__ om de velden te herinitialiseren bij de start van een nieuwe episode

De bedoeling is nu dat de agent via het __Q-learning algoritme__ leert om de juiste acties te nemen in de omgeving. De verschillende stappen in het Q-learning algoritme gaan als volgt :

---
1. Initialiseer een $Q-table$ : elk state-action pair krijgt initieel waarde 0
2. Implementeer de episolon-greedy actie selectie strategie
3. Voer de gekozen actie $a$ uit in toestand $s$ en observeer de nieuwe state $s_{new}$ en de immediate reward $r$
4. Gebruik bovenstaande ervaring om de Q-value Q(s,a) te updaten :

   $$Q(s,a) \leftarrow  Q(s,a) + \alpha (r + \gamma max_bQ(s_{new},b) - Q(s,a)) $$

5. Herhaal stap 3,4,5 nu voor toestand $s_{new}$ totdat een eindtoestand bereikt wordt
6. Herhaal bovenstaande stappen 3,4,5 voor voldoende aantal episodes
---

### 4. Simulatie in GYM
GYM is een Python API waarin single agent RL problemen kunnen gesimuleerd worden. De module bevat verschillende typische problemen waarop je je RL algoritmen kan testen. Je kan ook zelf nieuwe problemen hierin formuleren. Het bovenstaande probleem vind je eveneens terug in de gym omgeving onde de naam [Frozen Lake problem](https://gymnasium.farama.org/environments/toy_text/frozen_lake/). Het probleem is daar als volgt beschreven:

- de 4x4 versie heeft 16 toestanden $(0,1, \ldots 15 )$ en laad je in als volgt: 
```
gym.make('FrozenLake-v1', map_name="4x4", is_slippery=False, render_mode="rgb_array")
```

- de actieruimte heeft als shape $(1,)$ waarbij:
    - LEFT = 0 : move left
    - DOWN = 1 : move down
    - RIGHT = 2 : move right
    - UP = 3 : move up
- de rewardfunctie die hier gekozen werd
    - +1 bij het bereiken van de goal
    - -1 bij het doorzakken in het ijs
    - 0 elders

Het voordeel van de gym omgeving is onder andere dat je de agent policy visueel kan bekijken. Onderstaande code laat de beweging van een random agent in de frozen lake zien.

Je kan in deze code de random actie selectie wijzigen door je eigen geleerd policy en visueel waarnemen of je een optimale policy geleerd hebt.

````python
import numpy as np
import matplotlib.pyplot as plt
import gymnasium as gym
from gymnasium.envs.toy_text.frozen_lake import LEFT, RIGHT, DOWN, UP
from IPython.display import clear_output


def random_exploration(env, num_step):
    # Reset the environment to its initial state and get the initial observation (initial state)
    observation, info = env.reset(seed=2023)
    
    # Simulate the agent's actions for num_step time steps
    rewards = []
    for _ in range(num_step):
        # Choose a random action from the action space
        action = env.action_space.sample()
        ##### WIJZIG BOVENSTAANDE LIJN ZODAT JE Q-LEARNING AGENT INTELLIGENT ACTIES NEEMT
        print(action)

        # Take the chosen action and observe the resulting state, reward, and termination status
        observation, reward, terminated, truncated, info = env.step(action)
        rewards.append(reward)

        # If the episode is terminated, reset the environment to the start cell
        if terminated:
            observation, info = env.reset()

        # Display the current state of the environment
        clear_output(wait=True)
        plt.imshow(env.render())
        plt.show()
        
    return rewards

# Create the FrozenLake environment with specific settings
env = gym.make('FrozenLake-v1', map_name="4x4", is_slippery=False, render_mode="rgb_array")
rewards = random_exploration(env, 200)
env.close()
print(f"Total reward: {np.round(np.sum(rewards), 2)}")
````

### 5. Uitbreiding

Stel dat onze omgeving onderhevig is aan sterk oprukkend wind (is_slippery=True in de gym Frozen Lake versie), waardoor het nemen van acties bemoeilijkt wordt in deze omgeving. Concreet komt het neer op het feit dat bij het nemen van een actie er slechts $\frac{2}{3}$de kans is dat de agent terechtkomt in de cel waar hij naar toe wou, en $\frac{1}{3}$de kans op het afwijken hiervan. Deze afwijking houdt in dat de agent terechtkomt in de cel die zich onder de target cel bevindt (indien deze cel bestaat, indien niet dan blijft de agent in zijn oorspronkelijke cel).
